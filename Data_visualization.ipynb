## Dataset Overview

> This dataset includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area. There are 183412 trips and 16 features in the dataset. Most variables are numeric in nature, but the features start_station_name, end_station_name, user_type and member_gender are qualitative nominal variables.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import time

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('fordgobike_clean.csv')
```

## (Visualization 1)

> The top 10 stations are 'Market St at 10th St','San Francisco Caltrain Station 2  (Townsend St at 4th St)', 'Berry St at 4th St', 'Montgomery St BART Station (Market St at 2nd St)', 'Powell St BART Station (Market St at 4th St)', 'San Francisco Caltrain (Townsend St at 4th St)', 'San Francisco Ferry Building (Harry Bridges Plaza)', 'Howard St at Beale St', 'Steuart St at Market St', 'Powell St BART Station (Market St at 5th St)'

```python
df['start_station_name'].value_counts()[:10]
```

```python
# Subset the dataset and keep the top 10 locations with high trips

top10_loc = ['Market St at 10th St','San Francisco Caltrain Station 2  (Townsend St at 4th St)', 'Berry St at 4th St', 
        'Montgomery St BART Station (Market St at 2nd St)', 'Powell St BART Station (Market St at 4th St)',
        'San Francisco Caltrain (Townsend St at 4th St)', 'San Francisco Ferry Building (Harry Bridges Plaza)',
        'Howard St at Beale St', 'Steuart St at Market St', 'Powell St BART Station (Market St at 5th St)']
df_top10 = df.loc[df['start_station_name'].isin(top10_loc)]
```

```python
# top 10 station with the most trip

stat10_counts = df_top10['start_station_name'].value_counts()
order10 = stat10_counts.index

plt.figure(figsize=(8,7))
base_color = sb.color_palette()[0]

sb.countplot(data = df_top10, y = 'start_station_name', color = base_color, order = order10)
plt.xlabel('Number of trips')
plt.ylabel('Start Station Name')
plt.title('Top 10 Stations with the most trips')
plt.show()
```

## (Visualization 2)

> Users distribution by gender shows that there are over 120,000 trips taken by male, which is about 3 times larger than trips taken by female

```python
# show gender distribution

plt.figure(figsize = (7,7))
color = sb.color_palette()[0]
order = df.member_gender.value_counts().index

sb.countplot(data = df , x = 'member_gender' , color = color , order = order)
plt.title('User Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of trips')
plt.xticks(rotation = 45)
plt.show()
```

## (Visualization 3)
> About 90% of all trips was taken by subcribers, with customers making up for about 10% of the entire ride taken.


```python
# pie to show user_type

plt.figure(figsize=(7,7))
labels = ('Subscriber','Customer')
type_count = df['user_type'].value_counts()

plt.pie(type_count, startangle = 90, explode =(0.1, 0.1), labels = labels, autopct = '%1.0f%%', counterclock = False)
plt.title('Distribution by User Type')
plt.show()
```
